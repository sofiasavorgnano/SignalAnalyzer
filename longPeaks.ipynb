```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import glob
```

```python
file_path = glob.glob('/Users/sofiasavorgnano/Desktop/SignalAnalyzer/Longs/*.csv')
```

```python
time = np.linspace(0,200,501)
coe = np.array([])
inter = np.array([])
for j in range(10):
    mat = np.loadtxt(file_path[j],skiprows=19, delimiter=',')
    maxY = np.array([])
    for i in range(501):
        val = mat[:,i]
        maxY = np.append(maxY,np.max(val)) 

    # Linear fit 

    pars = np.polyfit(time,maxY,1)
    #ret = mx + q
    coe = np.append(coe,pars[0])
    inter = np.append(inter,pars[1])
    ret = pars[0]*time + pars[1]
    plt.figure()
    plt.plot(time,maxY,'b-')
    plt.plot(time,ret,'r-')
    plt.xlabel('time [s]')
    plt.ylabel('max amplitude [dB]')
    plt.grid()
print('m:',coe)
print('q:',inter)
```
